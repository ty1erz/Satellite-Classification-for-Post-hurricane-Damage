```python
trainDir=datasetPath+"/"+"train/"

validDir=datasetPath+"/"+"valid/"

testDir=datasetPath+"/"+"test/"
```

```python
def load_images(path,x,y):
    imgs = []
    label = []
    l1 = os.listdir(path)
    c=0
    for i in l1:
        l2 = os.listdir(path+'/'+i)
        for j in l2:

            img = Image.open(path+i+'/'+j).convert('RGB')
            img = img.resize(size=(x,y))
#             img = img.convert('L')
            imgs.append(np.array(img))
            label.append(i)
            del img
    return np.array(imgs),label
```

```python
x_train,y_train  = x_train,y_train=load_images(trainDir,128,128)
x_valid,y_valid  = x_valid,y_valid=load_images(validDir,128,128)

from sklearn.utils import shuffle

x_train, y_train = shuffle(x_train, y_train, random_state=0)
x_valid, y_valid = shuffle(x_valid, y_valid, random_state=0)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def display_random_samples(images, labels, num_samples=5):

    if num_samples > len(images):
        print("Number of samples requested exceeds the total number of images. Displaying all images instead.")
        num_samples = len(images)

    # Randomly select images and their corresponding labels
    random_indices = random.sample(range(len(images)), num_samples)
    selected_images = [images[i] for i in random_indices]
    selected_labels = [labels[i] for i in random_indices]

    # Set up the subplot grid
    fig, axes = plt.subplots(1, num_samples, figsize=(num_samples * 2, 4))

    if num_samples == 1:  # If only one sample, axes is not a list
        axes = [axes]

    for ax, img, label in zip(axes, selected_images, selected_labels):
        ax.imshow(img)
        ax.set_title(label)
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage
# Assuming `imgs` and `label` are obtained from the `load_images` function
display_random_samples(x_train, y_train, num_samples=10)
```

```python
t_train = pd.Series(y_train,dtype='category')
print(dict(enumerate(t_train.cat.categories)))
t_train = t_train.cat.codes

t_valid = pd.Series(y_valid,dtype='category')
print(dict(enumerate(t_valid.cat.categories)))
t_valid = t_valid.cat.codes
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
import tensorflow as tf
```

```python
modelAlex = models.Sequential()
# 1st layer (CONV + pool + batchnorm)
modelAlex .add(layers.Conv2D(filters= 96, kernel_size= (11,11), strides=(4,4), padding='valid', input_shape = (128,128,3)))
modelAlex.add(layers.Activation('relu'))
modelAlex.add(layers.MaxPool2D(pool_size=(3,3), strides=(2,2)))
modelAlex.add(layers.BatchNormalization())

# 2nd layer (CONV + pool + batchnorm)
modelAlex.add(layers.Conv2D(filters=256, kernel_size=(5,5), strides=(1,1), padding='same'))
modelAlex.add(layers.Activation('relu'))
modelAlex.add(layers.MaxPool2D(pool_size=(3,3), strides=(2,2), padding='valid'))
modelAlex.add(layers.BatchNormalization())

# layer 3 (CONV + batchnorm)
modelAlex.add(layers.Conv2D(filters=384, kernel_size=(3,3), strides=(1,1), padding='same'))
modelAlex.add(layers.Activation('relu'))
modelAlex.add(layers.BatchNormalization())

# layer 4 (CONV + batchnorm)
modelAlex.add(layers.Conv2D(filters=384, kernel_size=(3,3), strides=(1,1), padding='same'))
modelAlex.add(layers.Activation('relu'))
modelAlex.add(layers.BatchNormalization())

# layer 5 (CONV + batchnorm)
modelAlex.add(layers.Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), padding='same'))
modelAlex.add(layers.Flatten())
modelAlex.add(layers.Dense(4096)) #input_shape=(32,32,3,)))
modelAlex.add(layers.BatchNormalization())
modelAlex.add(layers.Activation('relu'))
# Add Dropout to prevent overfitting
modelAlex.add(layers.Dropout(0.4))

#2nd Fully Connected Layer
modelAlex.add(layers.Dense(4096))
modelAlex.add(layers.BatchNormalization())
modelAlex.add(layers.Activation('relu'))
#Add Dropout
modelAlex.add(layers.Dropout(0.4))

#3rd Fully Connected Layer
modelAlex.add(layers.Dense(1000))
modelAlex.add(layers.BatchNormalization())
modelAlex.add(layers.Activation('relu'))
#Add Dropout
modelAlex.add(layers.Dropout(0.4))

#Output Layer
modelAlex.add(layers.Dense(2))
modelAlex.add(layers.BatchNormalization())
modelAlex.add(layers.Activation('softmax'))
```

```python
modelAlex.summary()
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint

modelAlex.compile(loss="sparse_categorical_crossentropy", optimizer= 'adam', metrics=['accuracy'])
# es = EarlyStopping(monitor='val_loss', patience=3)
mc = ModelCheckpoint('best_model_alex.h5', monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)
historyAlex = modelAlex.fit(x_train,t_train,epochs=20,validation_data=(x_valid,t_valid),callbacks=[mc])
```

```python
import re

def extract_lat_long(file_name):
    match = re.search(r'([-+]?[0-9]*\.?[0-9]+)_([-+]?[0-9]*\.?[0-9]+)\.jpeg', file_name)
    if match:
        return float(match.group(1)), float(match.group(2))
    else:
        return None, None

def load_images_with_coordinates(path, x, y):
    imgs = []
    labels = []
    latitudes = []
    longitudes = []

    for label_dir in os.listdir(path):
        label_path = os.path.join(path, label_dir)
        if os.path.isdir(label_path):
            for img_file in os.listdir(label_path):
                if img_file.endswith('.jpeg'):
                    # Extract latitude and longitude
                    lat, long = extract_lat_long(img_file)

                    # Load the image
                    img_path = os.path.join(label_path, img_file)
                    img = Image.open(img_path).convert('RGB')
                    img = img.resize((x, y))
                    imgs.append(np.array(img))

                    # Append label and coordinates
                    labels.append(label_dir)
                    latitudes.append(lat)
                    longitudes.append(long)

                    del img

    return np.array(imgs), labels, latitudes, longitudes
```

```python
def extract_info_from_filename(file_name):
    # Split the filename and extract the required parts
    parts = file_name.split('_')
    if len(parts) >= 3:
        disaster_type = parts[0]  # e.g., 'hurricane' from 'hurricane-florence_00000186_pre_disaster.png'
        label = '_'.join(parts[-2:])  # e.g., 'pre_disaster' or 'post_disaster'
        return disaster_type, label
    else:
        return None, None

def load_images_real_data(path, x, y):
    imgs = []
    labels = []
    types = []

    for img_file in os.listdir(path):
            if img_file.endswith('.png'):
                # Extract disaster type and label from filename
                disaster_type, label = extract_info_from_filename(img_file)

                # Load the image
                img_path = os.path.join(path, img_file)
                img = Image.open(img_path).convert('RGB')
                img = img.resize((x, y))
                imgs.append(np.array(img))

                # Append label and coordinates
                labels.append(label)
                types.append(disaster_type)

                del img

    return np.array(imgs), np.array(labels), np.array(types)
```
